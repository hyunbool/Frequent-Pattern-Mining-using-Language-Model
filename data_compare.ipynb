## gpt

```python
gitemsets = pd.read_csv('gpt_candidates_small_001.csv')
```

```python
gcand = gitemsets['movie_candidates'].tolist()
```

```python
len(gcand)
```

```python
for i in range(len(gcand)):
    gcand[i] = gcand[i].replace("<", "")
    gcand[i] = gcand[i].replace("|", "")
    gcand[i] = gcand[i].replace("e", "")
    gcand[i] = gcand[i].replace("n", "")
    gcand[i] = gcand[i].replace("d", "")
    gcand[i] = gcand[i].replace("o", "")
    gcand[i] = gcand[i].replace("f", "")
    gcand[i] = gcand[i].replace("t", "")
    gcand[i] = gcand[i].replace("x", "")
    gcand[i] = gcand[i].replace(">", "")
    gcand[i] = gcand[i].replace("s", "")
    gcand[i] = gcand[i].replace("a", "")
    gcand[i] = gcand[i].replace("r", "")
```

```python
for i in range(len(gcand)):
    gcand[i] = gcand[i].split(" ")
```

```python
for i in range(len(gcand)):
    if '5' in gcand[i]:
        gcand[i].remove('5')
    if '25' in gcand[i]:
        gcand[i].remove('25')
    if '29' in gcand[i]:
        gcand[i].remove('29')
    if '119' in gcand[i]:
        gcand[i].remove('119')
    if '' in gcand[i]:
        gcand[i].remove('')
```

```python
for i in range(len(gcand)):
    gcand[i] = list(set(gcand[i]))
```

```python
gcand_list = []
for i in range(len(gcand)):
    if len(gcand[i]) > 1:
        gcand_list.append(gcand[i])
```

```python
gcand_list = [random.choice(gcand_list) for i in range(626)]
```

```python
for i in range(len(gcand_list)):
    for j in range(len(gcand_list[i])):
        gcand_list[i][j] = int(gcand_list[i][j])
```

```python
for i in range(len(gcand_list)):
    gcand_list[i] = sorted(gcand_list[i])
```

```python
for i in range(len(gcand_list)):
    for j in range(len(gcand_list[i])):
        gcand_list[i][j] = str(gcand_list[i][j])
```

## apriori & ngram 비교
- 정확도: apriori로 생성한 해당 부분집합 가지고 있는 n-gram 으로 생성한 아이템셋 갯수 / apriori로 생성한 해당 부분집합 가지고 있는 전체 데이터 갯수

```python
count = 0
litem = []

for i in range(len(cand)):
    for j in range(len(lcand_list)):
        if cand[i] == lcand_list[j]:
            count += 1
            litem.append(lcand_list[j])
    print(count)

prob = count/len(cand)
print("비율: %f" % prob)
```

```python
for i in range(len(litem)):
    litem[i] = str(litem[i])
```

```python
len(set(litem))
```

```python
test = []
for i in range(len(gcand_list)):
    test.append(str(gcand_list[i]))
```

```python
len(set(test))
```

```python
dup = list((set(test)) - (set(litem)))
dup
```

```python
for i in range(len(dup)):
    dup[i] = dup[i].replace("'", "")
    dup[i] = dup[i].replace("[", "")
    dup[i] = dup[i].replace("]", "")
```

```python
for i in range(len(dup)):
    dup[i] = dup[i].split(", ")
```

```python
for i in range(len(dup)):
    for j in range(len(dup[i])):
        dup[i][j] = int(dup[i][j])
```

```python
for i in range(len(dup)):
    dup[i] = sorted(dup[i])
```

# Apriori 계산

```python
te = TransactionEncoder()
te_result = te.fit(gcand_list).transform(gcand_list)
df = pd.DataFrame(te_result, columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```
