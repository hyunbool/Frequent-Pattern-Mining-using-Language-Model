```python
import pandas as pd
import pickle
```

```python
with open('test_movies_re.txt', 'rb') as f:
    movies = pickle.load(f)
```

```python
for i in range(len(movies)):
    for j in range(len(movies[i])):
        movies[i][j] = str(movies[i][j])
```

```python
movies
```

```python
movies_list = []
for i in range(len(movies)):
    movies_list.append(' '.join(movies[i]))
```

```python
movies_list
```

```python
whole = ""
for i in range(len(movies)):
    whole += "<|startoftext|>"
    whole += str(movies_list[i])
    whole += "<|endoftext|>"
```

```python
whole
```

```python
with open('train.txt', 'w', encoding='utf-8') as file:
    file.writelines(whole)
```
