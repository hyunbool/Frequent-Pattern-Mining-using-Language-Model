```python
import pandas as pd
import pickle
```

```python
with open('splitted_movies.txt', 'rb') as f:
    movies = pickle.load(f)
```

```python
movies
```

```python
splitted_movies = []
for i in range(len(movies)):
    if len(movies[i]) > 1:
        splitted_movies.append(movies[i])
    
```

```python
len(splitted_movies)
```

```python
df = pd.DataFrame(columns=['subset'])
location = 0

def powerset(s, df, location):
    x = len(s)
    for i in range(1 << x):
        if i == 0: continue #공집합은 빼기
        df.loc[location] = str([s[j] for j in range(x) if (i & (1 << j))])
        location += 1
        if location % 100 == 0:
            print("%d번째" % location)
    return df, location


# In[34]:


for i in range(len(splitted_movies)):
    df, location = powerset(splitted_movies[i], df, location)

# In[ ]:


df.to_csv('subset_test.csv')
```
